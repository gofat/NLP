```python
words_tagged = nltk.corpus.treebank.tagged_words(tagset='universal')
```

```python
# most common POS before NOUN
word_tag_pairs = nltk.bigrams(words_tagged)
noun_preceders = [a[1] for (a, b) in word_tag_pairs if b[1] == 'NOUN']
fdist = nltk.FreqDist(noun_preceders)
fdist.most_common() # ADP	adposition	on, of, at, with, by, into, under
```

```python
word_tag_pairs = nltk.bigrams(words_tagged)
verb_preceders = [a[0] for (a, b) in word_tag_pairs if b[1] == 'VERB']
fdist = nltk.FreqDist(verb_preceders)
fdist.most_common() # ADP	adposition	on, of, at, with, by, into, under
```

```python
cfd1 = nltk.ConditionalFreqDist(words_tagged)
```

```python
cfd1['yield'].most_common()
```
